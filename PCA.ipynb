```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Saumya\Downloads\earthquake\earthquakes.csv")
```

```python
numerical_columns = ['magnitude', 'latitude', 'longitude', 'depth']
```

```python
df_numerical = df[numerical_columns].dropna()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numerical)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

```python
print("PCA Results:")
print(df_pca.head())
print(f"Explained variance ratios: {explained_variance}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], c='blue', edgecolor='k', s=50)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Earthquake Data')
plt.grid(True)
plt.show()
```
